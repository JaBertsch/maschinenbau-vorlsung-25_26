## Aufgabe 1 ⭐ – Fläche eines Rechtecks

**Kontext:** Einstieg – einfache Funktion mit zwei Parametern und `return`.

**Aufgabe:** Schreibe die Funktion `flaeche(breite, hoehe)`, die die Fläche eines
Rechtecks zurückgibt.

**Erwartete Ausgabe:**
```
flaeche(3, 4) → 12
flaeche(2.5, 4) → 10.0
```


````python
# TODO: Implementiere die Funktion flaeche (aktuell liefert sie immer 0).
def flaeche(breite, hoehe):
    return breite * hoehe
print(flaeche(3, 4))
print(flaeche(2.5, 4))
````

````python
# ▶️ Selbstkontrolle Aufgabe 1
try:
    assert flaeche(3, 4) == 12, f"Erwartet 12, bekommen {flaeche(3, 4)}"
    assert flaeche(2.5, 4) == 10.0, f"Erwartet 10.0, bekommen {flaeche(2.5, 4)}"
    assert flaeche(0, 100) == 0
    print("✅ Aufgabe 1 gelöst.")
except AssertionError as e:
    print(f"❌ Noch nicht ganz: {e}")
except (NameError, TypeError):
    print("❌ Funktion flaeche noch nicht korrekt definiert.")
````

## Aufgabe 2 ⭐ – Drehzahl mit Default-Übersetzung

**Kontext:** An einem Getriebe wird eine Antriebsdrehzahl über eine Übersetzung in eine
Abtriebsdrehzahl umgerechnet: `n_ab = n_an * uebersetzung`.

**Aufgabe:** Schreibe `abtriebsdrehzahl(n_an, uebersetzung=1.0)`. Wenn keine Übersetzung
übergeben wird, soll die Antriebsdrehzahl unverändert zurückgegeben werden.

**Erwartete Ausgabe:**
```
abtriebsdrehzahl(1500) → 1500
abtriebsdrehzahl(1500, 0.5) → 750.0
```


````python
# TODO: Setze den Default-Parameter korrekt und implementiere die Formel n_an * uebersetzung.
def abtriebsdrehzahl(n_an, uebersetzung=1.0):
    return n_an * uebersetzung

print(abtriebsdrehzahl(1500))
print(abtriebsdrehzahl(1500, 0.5))
````

````python
# ▶️ Selbstkontrolle Aufgabe 2
try:
    assert abtriebsdrehzahl(1500) == 1500
    assert abtriebsdrehzahl(1500, 0.5) == 750.0
    assert abtriebsdrehzahl(3000, 2) == 6000
    print("✅ Aufgabe 2 gelöst.")
except AssertionError as e:
    print(f"❌ Noch nicht ganz: {e}")
except (NameError, TypeError):
    print("❌ Funktion abtriebsdrehzahl noch nicht korrekt definiert.")
````

## Aufgabe 4 ⭐⭐ – Statistik mit `*args`

**Kontext:** Du willst aus beliebig vielen Messwerten schnell eine Mini-Statistik ziehen.

**Aufgabe:** Schreibe `statistik(*werte)`, die ein Dictionary zurückgibt mit den
Schlüsseln `"anzahl"`, `"summe"`, `"mittelwert"`. Wenn **keine** Werte übergeben werden,
gib `{"anzahl": 0, "summe": 0, "mittelwert": 0}` zurück.

**Erwartete Ausgabe:**
```
statistik(2, 4, 6) → {"anzahl": 3, "summe": 12, "mittelwert": 4.0}
statistik() → {"anzahl": 0, "summe": 0, "mittelwert": 0}
```

**🗺️ Programmablaufplan (PAP):** [diagramme/aufgabe_4_pap.mmd](diagramme/aufgabe_4_pap.mmd)


````python
from IPython.display import Markdown, display
with open("diagramme/aufgabe_4_pap.mmd", encoding="utf-8") as f:
    display(Markdown(f"```mermaid\n{f.read()}\n```"))
````

````python
# TODO: Implementiere statistik mit *args.
def statistik(*werte):
    if len(werte) == 0:
        return {"anzahl": 0, "summe": 0, "mittelwert": 0}
    # TODO: anzahl, summe, mittelwert korrekt berechnen
    return {
        "anzahl": 0,
        "summe": 0,
        "mittelwert": 0,
    }

print(statistik(2, 4, 6))
print(statistik())
````

````python
# ▶️ Selbstkontrolle Aufgabe 4
try:
    r = statistik(2, 4, 6)
    assert r == {"anzahl": 3, "summe": 12, "mittelwert": 4.0}, f"Falsch: {r}"
    assert statistik() == {"anzahl": 0, "summe": 0, "mittelwert": 0}
    r2 = statistik(10, 20)
    assert r2["anzahl"] == 2 and r2["summe"] == 30 and r2["mittelwert"] == 15.0
    print("✅ Aufgabe 4 gelöst.")
except AssertionError as e:
    print(f"❌ Noch nicht ganz: {e}")
except (NameError, TypeError) as e:
    print(f"❌ Funktion statistik noch nicht korrekt definiert: {e}")
````

## Aufgabe 5 ⭐⭐ – Bauteil-Template mit `**kwargs`

**Kontext:** Du hast eine Textvorlage [`cad_template_technisch.txt`](cad_template_technisch.txt)
mit Platzhaltern wie `{BEZEICHNUNG}`, `{NUMMER}`, `{MATERIAL}`, `{ABMESSUNGEN}`,
`{BESCHREIBUNG}`, `{BEARBEITER}`, `{DATUM}`, `{STATUS}`.

**Aufgabe:** Schreibe `fuelle_template(pfad, **felder)`, die:
1. den Templatetext aus der Datei unter `pfad` lädt,
2. **jeden** in `felder` übergebenen Schlüssel `k` durch seinen Wert ersetzt,
   indem `{k.upper()}` (also z. B. `{BEZEICHNUNG}`) im Text durch den Wert getauscht wird,
3. den fertigen Text zurückgibt.

Tipp: `text.replace("{BEZEICHNUNG}", "Zahnrad")` ersetzt den Platzhalter.

**Erwartete Ausgabe (Auszug):**
```
Bezeichnung: Zahnrad
Bauteil-Nummer: ZR-042
Material: Stahl
```

**🗺️ Programmablaufplan (PAP):** [diagramme/aufgabe_5_pap.mmd](diagramme/aufgabe_5_pap.mmd)


````python
with open("diagramme/aufgabe_5_pap.mmd", encoding="utf-8") as f:
    display(Markdown(f"```mermaid\n{f.read()}\n```"))
````

````python
# TODO: Implementiere fuelle_template mit **kwargs.
def fuelle_template(pfad, **felder):
    with open(pfad, encoding="utf-8") as f:
        text = f.read()
    # TODO: Für jedes k, v in felder.items() das {K} (Großbuchstaben!) im Text ersetzen.
    #       Aktuell wird der Text unverändert zurückgegeben.
    return text


ausgabe = fuelle_template(
    "cad_template_technisch.txt",
    bezeichnung="Zahnrad",
    nummer="ZR-042",
    material="Stahl",
    abmessungen="Ø 50 mm x 12 mm",
    beschreibung="Geradverzahntes Stirnrad, 42 Zähne.",
    bearbeiter="Studentin A",
    datum="2025-11-15",
    status="Freigegeben",
)
print(ausgabe)
````

````python
# ▶️ Selbstkontrolle Aufgabe 5
try:
    assert "Bezeichnung: Zahnrad" in ausgabe, "Bezeichnung fehlt oder falsch ersetzt."
    assert "Bauteil-Nummer: ZR-042" in ausgabe
    assert "Material: Stahl" in ausgabe
    assert "{BEZEICHNUNG}" not in ausgabe, "Platzhalter {BEZEICHNUNG} wurde nicht ersetzt."
    assert "{STATUS}" not in ausgabe, "Platzhalter {STATUS} wurde nicht ersetzt."
    print("✅ Aufgabe 5 gelöst – dein Template-Filler funktioniert.")
except AssertionError as e:
    print(f"❌ Noch nicht ganz: {e}")
except (NameError, TypeError) as e:
    print(f"❌ Funktion fuelle_template noch nicht korrekt definiert: {e}")
except FileNotFoundError:
    print("❌ cad_template_technisch.txt nicht gefunden – Notebook aus dem Ordner V11 ausführen.")
````
