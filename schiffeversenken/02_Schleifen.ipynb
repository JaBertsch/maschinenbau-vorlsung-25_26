## Block 1: `for`-Schleife (45 min)

### `range()` â€” Zahlenfolgen erzeugen

Die `for`-Schleife wiederholt Code fĂĽr jedes Element einer Sequenz. Mit `range()` erzeugst du eine Zahlenfolge:

| Aufruf | Erzeugt | Beschreibung |
|--------|---------|-------------|
| `range(5)` | 0, 1, 2, 3, 4 | 5 Zahlen, startet bei 0 |
| `range(1, 6)` | 1, 2, 3, 4, 5 | Von 1 bis 5 (Endwert exklusive) |

**Merke:** Der Endwert ist immer **exklusive** â€” `range(1, 6)` liefert 1 bis 5, nicht bis 6.

````python
# range(n) â€” startet bei 0
for i in range(5):
    print(i)
````

````python
# range(start, stop) â€” Bereich festlegen
for i in range(1, 6):
    print(f"Zeile {i}")
````

### Iteration ĂĽber Sequenzen

Eine `for`-Schleife kann auch direkt ĂĽber die Zeichen eines Strings iterieren:

````python
# Ăśber einen String iterieren
for buchstabe in "ABCDE":
    print(buchstabe)
````

### `enumerate()` â€” Index und Wert gleichzeitig

Manchmal brauchst du sowohl den **Index** (Position) als auch den **Wert**. DafĂĽr gibt es `enumerate()`:

````python
# enumerate() liefert Index und Wert als Paar
spalten = "ABCDE"

for index, buchstabe in enumerate(spalten):
    print(f"Spalte {index}: {buchstabe}")
````

### Verschachtelte Schleifen

FĂĽr 2D-Ausgaben (z.B. ein Raster) braucht man eine **Ă¤uĂźere Schleife** (Zeilen) und eine **innere Schleife** (Spalten):

```
Zeile 1: Spalte A, Spalte B, Spalte C, ...
Zeile 2: Spalte A, Spalte B, Spalte C, ...
...
```

Die innere Schleife lĂ¤uft fĂĽr **jede Iteration** der Ă¤uĂźeren Schleife komplett durch.

````python
# Verschachtelte Schleife: 3Ă—3 Raster
for zeile in range(3):
    for spalte in range(3):
        print("+", end=" ")  # end=" " verhindert den Zeilenumbruch
    print()  # Zeilenumbruch nach jeder Zeile
````

````python
# Spielfeld-Kontext: SpaltenkĂ¶pfe + Raster
# SpaltenkĂ¶pfe
print("  ", end="")
for buchstabe in "ABCDEF":
    print(f" {buchstabe}", end="")
print()  # Zeilenumbruch

# Zeilen mit Zeilennummern
for zeile in range(1, 5):  # nur 3 Zeilen als Demo
    print(f"{zeile} ", end="")
    for spalte in range(6):
        print(" ~", end="")
    print()
````

### `print()` mit `end`-Parameter

StandardmĂ¤Ăźig fĂĽgt `print()` am Ende einen Zeilenumbruch ein. Mit `end` kann man das Ă¤ndern:

- `print("Hallo", end=" ")` â†’ kein Zeilenumbruch, stattdessen ein Leerzeichen
- `print()` â†’ nur ein Zeilenumbruch (nĂĽtzlich nach einer Zeile im Raster)

## Block 2: `while`-Schleife (45 min)

### `for` vs. `while`

| | `for` | `while` |
|---|-------|--------|
| **Wann?** | Anzahl der DurchlĂ¤ufe ist **bekannt** | Abbruchbedingung entscheidet |
| **Beispiel** | "Gib 5 Zeilen aus" | "Frage bis gĂĽltige Eingabe" |
| **Syntax** | `for x in sequenz:` | `while bedingung:` |

````python
# while-Schleife: ZĂ¤hler von 1 bis 5
zaehler = 1

while zaehler <= 5:
    print(f"Durchlauf {zaehler}")
    zaehler = zaehler * 2  # Wichtig: ZĂ¤hler erhĂ¶hen!
````

### Endlosschleifen erkennen und vermeiden

Eine `while`-Schleife lĂ¤uft **endlos**, wenn die Bedingung nie `False` wird:

```python
# ACHTUNG: Endlosschleife!
zaehler = 1
while zaehler <= 5:
    print(zaehler)
    # zaehler wird nie erhĂ¶ht â†’ Bedingung bleibt True
```

**Faustregel:** In jeder `while`-Schleife muss sich etwas Ă¤ndern, das die Bedingung beeinflusst.

Falls eine Endlosschleife passiert: In Jupyter Notebook den **Kernel unterbrechen** (Stopp-Button â¬ś).

### `break` â€” Schleife vorzeitig beenden

`break` beendet die Schleife sofort, egal ob die Bedingung noch `True` ist:

````python
# break: Abbruch wenn gewĂĽnschter Wert gefunden
for zahl in range(1, 100):
    if zahl * zahl > 50:
        print(f"{zahl} ist die erste Zahl deren Quadrat > 50 ist")
        break
````

### `continue` â€” Durchlauf ĂĽberspringen

`continue` ĂĽberspringt den Rest des aktuellen Durchlaufs und springt zum nĂ¤chsten:

````python
# continue: Ungerade Zahlen ĂĽberspringen
for zahl in range(1, 11):
    if zahl % 2 != 0:  # % ist der Modulo-Operator (Rest der Division)
        continue
    print(f"{zahl} ist gerade")
````

### Muster: Eingabeschleife mit `while True` + `break`

Ein hĂ¤ufiges Muster: Frage den Nutzer wiederholt, bis eine gĂĽltige Eingabe kommt.

```python
while True:
    eingabe = input("Eingabe: ")
    if eingabe_ist_gueltig:
        break
    print("UngĂĽltige Eingabe, versuche es erneut.")
```

Das `while True` lĂ¤uft endlos â€” der `break` in der `if`-Bedingung ist der einzige Ausgang.

````python
# Beispiel: Passwort-Eingabe (simuliert mit fester PrĂĽfung)
# Zum Testen: gib "geheim" ein
passwort = "geheim"   
while True:
    passwort = input("Passwort eingeben: ")
    if passwort == "geheim":
        print("Zugang gewĂ¤hrt!")
        break
    print("Falsches Passwort, nochmal.")
````
